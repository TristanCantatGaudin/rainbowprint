```python
from rainbowprint.functions import rainbowprint
rainbowprint('This is a rainbow')
rainbowprint('This is another rainbow','coolwarm')
```

```python
import matplotlib.pyplot as plt
for cmap_id in plt.colormaps():
    string_to_print = cmap_id+' '
    while len(string_to_print) < 30:
        string_to_print+='='
    rainbowprint(string_to_print,cmap_id)
```
